## Data Cleaning

```python
# Read CSV files into a pandas DataFrame
hawaii_measurements_df = pd.read_csv("Resources/hawaii_measurements.csv", dtype=object)
hawaii_stations_df = pd.read_csv("Resources/hawaii_stations.csv", dtype=object)
```

```python
#Determine if there are any null or NaN for each df
hawaii_measurements_df.isnull().sum()
```

```python
# Drop the NaN from the hawaii_measurements_df 
hawaii_measurements_clean = hawaii_measurements_df.dropna(axis=0)
hawaii_measurements_clean.head(10)
```

```python
hawaii_measurements_clean.reset_index()
```

```python
hawaii_measurements_clean.isnull().sum()
```

```python
# Save each clane df to csv file
hawaii_measurements_clean.to_csv('data/clean_hawaii_measurements.csv', encoding='utf-8', index=False)
hawaii_stations_df.to_csv('data/clean_hawaii_stations.csv',encoding='utf-8', index=False)
```
